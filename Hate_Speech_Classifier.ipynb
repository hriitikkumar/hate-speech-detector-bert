```python
# Install packages (if needed — skip if already installed)
!pip install -q transformers torch

# Step 1: Imports
import pandas as pd
import re, string
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizerFast, BertForSequenceClassification, AdamW
from sklearn.model_selection import train_test_split

# Step 2: Load and clean data
df = pd.read_csv('labeled_data.csv')
df = df[['tweet', 'class']]
label_map = {0: 'Hate Speech', 1: 'Offensive', 2: 'Neither'}
df['label'] = df['class'].map(label_map)

def clean_tweet(text):
    text = text.lower()
    text = re.sub(r'\brt\b', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'@\w+|#\w+', '', text)
    text = re.sub(f"[{re.escape(string.punctuation)}]", " ", text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_tweet'] = df['tweet'].apply(clean_tweet)

# Step 3: Train-validation split
train_df, val_df = train_test_split(
    df[['clean_tweet', 'class']],
    test_size=0.15,
    stratify=df['class'],
    random_state=42
)

# Step 4: Tokenization
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')
max_len = 64

def tokenize_batch(texts):
    return tokenizer(
        texts.tolist(),
        padding='max_length',
        truncation=True,
        max_length=max_len,
        return_tensors='pt'
    )

X_train = tokenize_batch(train_df['clean_tweet'])
y_train = train_df['class'].values

X_val = tokenize_batch(val_df['clean_tweet'])
y_val = val_df['class'].values

# Step 5: Dataset class
class TweetDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(int(self.labels[idx]), dtype=torch.long)
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = TweetDataset(X_train, y_train)
val_dataset = TweetDataset(X_val, y_val)

# Step 6: DataLoaders
train_dataloader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=32)

# Step 7: Model setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3)
model.to(device)

optimizer = AdamW(model.parameters(), lr=2e-5)

# Step 8: Training loop
epochs = 3

for epoch in range(epochs):
    print(f"\nEpoch {epoch + 1}/{epochs}")
    model.train()
    total_loss = 0

    for batch in train_dataloader:
        b_input_ids = batch['input_ids'].to(device)
        b_attention_mask = batch['attention_mask'].to(device)
        b_labels = batch['labels'].to(device)

        optimizer.zero_grad()
        outputs = model(
            input_ids=b_input_ids,
            attention_mask=b_attention_mask,
            labels=b_labels
        )

        loss = outputs.loss
        total_loss += loss.item()
        loss.backward()
        optimizer.step()

    avg_train_loss = total_loss / len(train_dataloader)
    print(f"Average training loss: {avg_train_loss:.4f}")
```

```python
def predict(text):
    model.eval()
    tokens = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
    tokens = {k: v.to(device) for k, v in tokens.items()}

    with torch.no_grad():
        output = model(**tokens)
        pred_label = torch.argmax(output.logits, dim=1).item()

    print(f"Prediction: {label_map[pred_label]}")

# Example usage
predict("seeing your kind around freely in this world makes me sick")  # 🔥 Try this with different inputs
```
